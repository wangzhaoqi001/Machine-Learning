### Quiz 1.a
In this example, the prevalence is 5%. I assumed that the number of people is X, so we can know the equation below:
\begin{align}
True Positive+False Negative&= 0.05*X
\end{align}

And this test has 90% sensitvity and 95% specificity, so we can know equations below:
\begin{align}
\\ \frac{TruePositive}{TruePositive+FalseNegative}&=0.9 \\\\
\frac{TrueNegative}{TrueNegative+FalsePositive}&=0.95 \\
\end{align}

Then, we can know:
\begin{align}
TruePositive&=9*FalseNegative \\
TrueNegative&=19*FalsePositive \\
\end{align}

According to the first equation, we can get:
\begin{align}
TruePositive+FalseNegative&= 0.05*X\\
9*FalseNegative+FalseNegative&= 0.05*X\\
FalseNegative=0.005*X\\
\end{align}

Then:
\begin{align}
TrueNegative+FalsePositive&= 0.95*X\\
19*FalsePositive+FalsePositive&= 0.95*X\\
FalseNegative=0.0475*X\\
\end{align}

Now, we can calculate the positive predictive valueï¼š
\begin{align}
PPV&=\frac{TruePositive}{TruePositive+FalsePositive}\\
&=\frac{9*FalseNegative}{9*FalseNegative+FalsePositive}\\
&=\frac{9*0.005*X}{9*0.005*X+0.0475*X}\\
&=\frac{0.045*X}{0.0925*X}\\
&\approx 49\% \\
\end{align}

According to the confusion matrix, we can calculate the accuracy, precision, recall and F1-score.
\begin{align}
Accuracy=&\frac{TruePositive+TrueNegative}{TruePositive+TrueNegative+FalsePositive+FalseNegative}\\
Precision=&\frac{TruePositive}{TruePositive+FalsePositive}\\
Recall=&\frac{TruePositive}{TruePositive+FalseNegative}\\
F1-score=&\frac{2*Precision*Recall}{Precision+Recall}\\
\end{align}

### I chose **Precision** as the most important metrics to evaluate the result of models. Because I think people usually can't accept the normal email been filtered as a spam email. Compared with missing some important messages, reading some spam emails is not bad.

### We can see that Random Forest, Logistic Regression, Naive Bayes get 100% Precision. But the Recall of the Logistic Regression is higher than the two others. So I think the **Logistic Regression** gets the best result of filtering spam emails.
